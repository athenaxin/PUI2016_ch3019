```python
import numpy as np
import pandas as pd
import scipy.stats as st
import matplotlib.pylab as plt

%matplotlib inline
```

```python
D = [0.2, 0.4, 0.8, 1.6, 3.2, 6.4]

ax = plt.figure(figsize = (15, 12)).add_subplot(111)
for d in D:
    x = np.linspace(-6, 6, 10000)
    s = np.sqrt(d)
    ax.plot(x, st.norm.pdf(x, 0, s), lw = 2.5, label = '$\mu = 0, \sigma^2 = {}$'.format(d))

ax.set_xlim(-5.05, 5.05)
ax.set_ylim(-0.01, 1.01)
plt.legend(loc = 1, fontsize = 'xx-large', markerfirst = False)
ax.text(-4.5, 0.9, '$\\varphi_{\mu,\sigma^2}(X)=\\frac{1}{\sqrt{2\sigma^2\pi}}e^{-\\frac{(x-\mu)^2}{2\sigma^2}}$', 
        fontsize = 25, style = 'italic', bbox = {'facecolor':'white', 'alpha':1, 'pad':10})

ax.set_xticks(np.arange(-5, 5.1, 1))
ax.set_yticks(np.arange(0, 1.01, 0.1))
plt.grid(True)

ax.set_xlabel('$X$', fontsize = 20)
ax.set_ylabel('$\\varphi_{\mu, \sigma^2}(X)$', fontsize = 22)

plt.suptitle('  Norm Distrbution', fontsize = 24, fontweight = 'bold')
ax.set_title('Probability Density Function', fontsize = 22, fontweight = 'bold')

plt.show()
```
