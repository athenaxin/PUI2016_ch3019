## 4. Now fit the data with a line.

```python
def line(x, a, b):
    return a * x + b
```

```python
males = np.array([0.] + [allmales[k]['Median income'][1] for k in race])
females = np.array([0.] + [allfemales[k]['Median income'][1] for k in race])
```

```python
beta = np.polyfit(males, females, 1)
```

```python
pl.figure(figsize = (10, 10))

for k in range(5):
    pl.scatter(males[k + 1], females[k + 1], c = pcolors[k], s = 200, label = race[k])

x = np.linspace(10000, 45000, 10)
pl.plot(x, line(x, beta[0], beta[1]), label = 'polyfit')
pl.xlabel("male", fontsize = 20)
pl.ylabel("female", fontsize = 20)
pl.title("Total Median Income by Race", fontsize = 25)
pl.legend(loc = 'best', scatterpoints = 1)
```

### Figure 2: Total Median Income of female vs male by Race and Polyfit Line

```python
cor = stats.spearmanr(males, females)
beta1 = cor[0] * females.std() / males.std()
beta0 = females.mean() - beta1 * males.mean()
beta0, beta1
```

```python
pl.figure(figsize = (10, 10))

for k in range(5):
    pl.scatter(males[k + 1], females[k + 1], c = pcolors[k], s = 200, label = race[k])

x = np.linspace(10000, 45000, 10)
pl.plot(x, line(x, beta[0], beta[1]), label = 'polyfit')
pl.plot(x, line(x, beta1, beta0), label = 'analytical')
pl.xlabel("male", fontsize = 20)
pl.ylabel("female", fontsize = 20)
pl.title("Total Median Income by Race", fontsize = 25)
pl.legend(loc = 'best', scatterpoints = 1)
```

### Figure 3: Total Median Income of female vs male by Race and Polyfit Line and Analytical Solution

## 5. this was done with THE TOTAL MEDIAN INCOME, let's now see if all forms of income lead to a different fit.

```python
males_incomes = np.array( \
    np.concatenate([allmales[k]['Median income'][:].values for k in race]).flatten())
females_incomes = np.array( \
    np.concatenate([allfemales[k]['Median income'][:].values for k in race]).flatten())

mask = ~(np.isnan(males_incomes) + np.isnan(females_incomes))

males_i = males_incomes[mask]
females_i = females_incomes[mask]
```

```python
data = pd.DataFrame({"female": females_i, "male": males_i})
```

```python
beta_i = np.polyfit(males_i, females_i, 1)

cor = stats.spearmanr(males_i, females_i)
beta1_i = cor[0] * females_i.std() / males_i.std()
beta0_i = females_i.mean() - beta1 * males_i.mean()

lm = smf.ols(formula = "female ~ male - 1", data = data).fit()
```

```python
pl.figure(figsize = (10, 10))

for k in range(5):
    pl.scatter(males[k + 1], females[k + 1], c = pcolors[k], s = 200, label = race[k])
pl.scatter(males_i, females_i, s = 20, alpha = 0.3, label = 'data')

x = np.linspace(0, 48000, 10)
pl.plot(x, line(x, beta_i[0], beta_i[1]), 'r', label = 'polyfit')
pl.plot(x, line(x, beta1_i, beta0_i), 'k', label = 'analytical')
pl.plot(x, line(x, lm.params[0], 0), 'b', label = 'regression through 0')
pl.xlabel("male", fontsize = 20)
pl.ylabel("female", fontsize = 20)
pl.title("Total Median Income", fontsize = 25)
pl.legend(loc = 'best', scatterpoints = 1)
```
